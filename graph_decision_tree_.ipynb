```python
import pandas as pd
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import pydotplus
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
```

```python
df=pd.read_csv('show.csv')
df
```

```python
d={'UK':0,'USA':1,'N':2}
df['Nationality']=df['Nationality'].map(d)
df
```

```python
m={'YES':1,'NO':0}
df['Go']=df['Go'].map(m)
df
```

```python
features=['Age','Experience','Rank','Nationality']
x=df[features]
```

```python
y=df['Go']
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
model=DecisionTreeClassifier()
model.fit(xtrain,ytrain)
```

```python
graphviz=tree.export_graphviz(model,out_file=None,feature_names=features)
```

```python
graph=pydotplus.graph_from_dot_data(graphviz)
```

```python
graph.write_png('mydecisintree.png')
```

```python
img=pltimg.imread('mydecisintree.png')
```

```python
imgplot=plt.imshow(img)
plt.show()
```
